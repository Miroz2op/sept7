```python
import numpy as np
import sklearn as sk
from sklearn.metrics import mean_squared_error, classification_report, silhouette_samples
from sklearn.linear_model import LinearRegression
```

```python
#Задание 1
y_true = [0,1,2,3]
y_pred = [8, 7, 6, 7]
print(mean_squared_error(y_true, y_pred))
```

```python
#Задание 2
y_true = [0,0,1,1,0,0,1,1,1]
y_pred = [0, 0, 0, 1, 0, 1, 0, 0, 0]
print(classification_report(y_true, y_pred, zero_division=False))
```

```python
#Задание 3
y = [0,0,1,1]
X = [(3, 1), (3, 3), (2, 3), (1, 4)]

sample_silhouette_values = silhouette_samples(X, y, metric='euclidean')
print('silhouette coefficients:', *sample_silhouette_values, sep='\n')

sample_silhouette_mean = sample_silhouette_values.mean()
print('silhouette mean: ', sample_silhouette_mean)
```

```python
#Задание 4
x = np.array([0, 1, 2]).reshape((-1, 1))
y = np.array([3, 1, 0])
model = LinearRegression().fit(x, y)
a = model.coef_[0]
b = model.intercept_
print(f'a = {a}\nb = {b}')
```
